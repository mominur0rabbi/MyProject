```python
import pandas as pd
```

```python
mainsub = pd.read_csv('sci_7.csv')
```

```python
Ban = pd.read_csv('Re_Ban2S.csv')
```

```python
Newmain = pd.merge(mainsub, Ban, how = 'outer')
```

```python
eng = pd.read_csv('Re_Engs.csv')
```

```python
Newmain1 = pd.merge(Newmain, eng, how = 'outer')
```

```python
ict = pd.read_csv('Re_ICTs.csv')
```

```python
Newmain2 = pd.merge(Newmain1, ict, how = 'outer')
Newmain2.head()
```

```python
phy = pd.read_csv('Re_Phy.csv')
Newmain3 = pd.merge(Newmain2, phy, how = 'outer')
Newmain3.head()
```

```python
che = pd.read_csv('Re_Chem.csv')
Newmain4 = pd.merge(Newmain3, che, how = 'outer')
Newmain4.head()
```

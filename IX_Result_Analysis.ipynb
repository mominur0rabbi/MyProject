```python
# import pandaa and Numpy
import pandas as pd
import numpy as np
```

```python
# Read csv files
df = pd.read_csv('X_Result.csv')
```

```python
df.head()
```

```python
# Sum Two Column, Make Grade point and Later Grade on new Sum
df['CheTTL'] = df['CheCQ']+df['Che20']
df['CheGP'] =np.where(df['CheTTL']>39, 5, np.where(df['CheTTL']>34, 4,np.where(df['CheTTL']>29, 3.5,np.where(df['CheTTL']>24, 3, np.where(df['CheTTL']>19, 2,0)))))
df['CheGL'] =np.where(df['CheGP']==5, 'A+', np.where(df['CheGP']==4, 'A',np.where(df['CheGP']==3.5, 'A-',np.where(df['CheGP']==3, 'B', np.where(df['CheGP']==2, 'C','F')))))
```

```python
df = df.round({'GPA': 2})
df.head()
```

```python
#Grand Total of All Subject Total
df['TTLMark'] = df['BANTTL']+df['EngTTL']+df['GMathTTL']+df['BGSTTL']+df['ReliTTL']+df['PhyTTL']+df['CheTTL']+df['BioTTL']+df['ICT25']+df['HMTTL']
```

```python
df['TTLGP'] = df['BANGP']+df['EngGP']+df['GMathGP']+df['BGSGP']+df['ReliGP']+df['PhyGP']+df['CheGP']+df['BioGP']+df['ICTGP']+df['HMGP']

```

```python
df['AGP']=df['TTLGP']/10
df['LG'] =np.where(df['AGP']>=5, 'A+', np.where(df['AGP']>=4, 'A',np.where(df['AGP']>=3.5, 'A-',np.where(df['AGP']>=3, 'B', np.where(df['AGP']>=2, 'C','F')))))
```

```python
df = df.sort_values(by=['TTLGP', 'TTLMark'], ascending=[False, False])
```

```python
# Add Serial Number
df['Position'] = range(1, 1+len(df))
```

```python
# Rearrange Columns 
df = df[['ID', 'BANCQ', 'BAN20','BanTTL', 'BanGP', 'BanGL',
         'ENGCQ', 'ENG20','EngTTL', 'EngGP', 'EngGL',
         'MathCQ',  'Math20','MathTTL', 'MathGP', 'MathGL',
         'BGSCQ', 'BGS20','BGSTTL',  'BGSGP', 'BGSGL',
         'ReliCQ', 'Reli20', 'ReliTTL', 'ReliGP',  'ReliGL',
         'PhyCQ', 'Phy20', 'PhyTTL', 'PhyGP', 'PhyGL',
         'CheCQ', 'Che20', 'CheTTL',  'CheGP',  'CheGL',
         'BioCQ', 'Bio20', 'BioTTL',  'BioGP', 'BioGL',
         'HMCQ', 'HM20', 'HMTTL', 'HMGP', 'HMGL',
         'TTLMark','Per', 'TTLGPA', 'GPA', 'LG', 'Position']]
```

```python
import pandas as pd
import os
```

```python
files = [file for file in os.listdir('./SecondTermXI/')]
all_data = pd.DataFrame()
for file in files:
    df = pd.read_csv("./SecondTermXI/"+file)
    all_data = pd.concat([all_data,df])
all_data.head()
```

```python
all_data.to_csv("math.csv", index = False)
```
